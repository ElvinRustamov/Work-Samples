```python
import pandas as pd
import numpy as np

# For visualization

import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels
```

```python
# Read Datased

df = pd.read_csv('Stores.csv')
```

```python
# Look at the Data

df.head()
```

```python
# Check if there is any missing value

df.isna().sum()
```

```python
# Informatin about data | With info()

df.info()
```

```python
# Describe the Data

df.describe()
```

```python
analysis_all = ["Items_Available", "Daily_Customer_Count", "Store_Area"]
color_all = ["red", "blue", "black"]
title_all = ["Store Sales vs Items Available", "Store Sales vs Daily Customer Count", "Store Sales vs Store Area"]
for i,j,k in zip(analysis_all,color_all, title_all):
    sns.lmplot(x ="Store_Sales" ,y= i, data = df, lowess = True, aspect = 3, line_kws = {'color':j}).set(title=k)
    plt.show()
```

```python
# Relation Between Daily_Customer_Count and Items_Available

df = df.sort_values(by = "Daily_Customer_Count")
sns.catplot(x = "Daily_Customer_Count",y = "Items_Available", data = df, aspect =4)
plt.xticks(rotation =90)
plt.show()
```

```python
# Store Sales Visualization

df = df.sort_values(by = "Daily_Customer_Count")
sns.violinplot(y = "Store_Sales", data = df)
plt.show()
```
