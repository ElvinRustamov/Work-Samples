Information (data) about Azerbaijan

```python
Azerbaijan_unemployment_analysis = unemployment_analysis.loc[unemployment_analysis['Country Name'] == 'Azerbaijan']
Azerbaijan_unemployment_analysis
```

Maximum and minimum unemployment rate in Azerbaijan between 1991 and 2021

```python
AZE_max_min = Azerbaijan_unemployment_analysis.melt(id_vars=['Country Name', 'Country Code'], var_name='Years', value_name='Rate %')
AZE_max_min
```

```python
Maximum_value = AZE_max_min.loc[AZE_max_min['Rate %'] == AZE_max_min['Rate %'].max(), ['Years', 'Rate %']]
Minimum_value = AZE_max_min.loc[AZE_max_min['Rate %'] == AZE_max_min['Rate %'].min(), ['Years', 'Rate %']]

print(f'Maximum Unemployment Rate: {Maximum_value.iloc[0, 1]} Year: {Maximum_value.iloc[0, 0]} ')
print(f'Minimum Unemployment Rate: {Minimum_value.iloc[0, 1]} Year: {Minimum_value.iloc[0, 0]} ')
```

```python
sns.lineplot(data=AZE_max_min, x='Years', y='Rate %');
plt.xticks(rotation=60);
plt.title('Unemployment rate in Azerbaijan between 1991 and 2021')
```

```python
rates = [f'{rate}%' for rate in list(AZE_max_min.loc[:, 'Rate %'])]
sns.barplot(data=AZE_max_min, x='Years', y='Rate %', label=rates);
plt.xticks(rotation=60);
plt.legend(bbox_to_anchor=(1, 1.05))
plt.title('Unemployment rate in Azerbaijan between 1991 and 2021');
```

Countries with the lowest and highest unemployment rate in the world in 2021.

```python
world_unemployment_rate_2021 = unemployment_analysis.loc[:, ['Country Name', 'Country Code', '2021']]
highest_unemployment_rate = unemployment_analysis.loc[unemployment_analysis['2021'] == unemployment_analysis['2021'].max()].melt(id_vars=['Country Name', 'Country Code'], var_name='Years', value_name='Rate %')
lowest_unemployment_rate = unemployment_analysis.loc[unemployment_analysis['2021'] == unemployment_analysis['2021'].min()].melt(id_vars=['Country Name', 'Country Code'], var_name='Years', value_name='Rate %')


sns.lineplot(data=highest_unemployment_rate, x='Years', y='Rate %', label='South Africa');
sns.lineplot(data=lowest_unemployment_rate, x='Years', y='Rate %', label='Qatar');
plt.xticks(rotation=60);
plt.title(label='Difference between countries which has the lowest and highest unemployment rate in the world in 2021.')
plt.legend();
```

```python
rates = [f'{rate}%' for rate in list(lowest_unemployment_rate.loc[:, 'Rate %'])]
sns.barplot(data=lowest_unemployment_rate, x='Years', y='Rate %', label=rates);
plt.xticks(rotation=60);
plt.legend(bbox_to_anchor=(1, 1.05))
plt.title('Unemployment rate in Qatar')
```

```python
rates = [f'{rate}%' for rate in list(highest_unemployment_rate.loc[:, 'Rate %'])]
sns.barplot(data=highest_unemployment_rate, x='Years', y='Rate %', label=rates);
plt.xticks(rotation=60);
plt.legend(bbox_to_anchor=(1, 1.05))
plt.title('Unemployment rate in South Africa')
```

Change in the average unemployment rate of all countries in the world (1991-2021)

```python
world_unemployment_rate = unemployment_analysis.melt(id_vars=['Country Name', 'Country Code'], var_name='Years', value_name='Rate %').groupby('Years')['Rate %'].agg('mean')
world_unemployment_rate = world_unemployment_rate.reset_index()

sns.lineplot(data=world_unemployment_rate, x='Years', y='Rate %');
plt.xticks(rotation=60);
plt.title('Change in the average unemployment rate of all countries in the world (1991-2021)');
```
