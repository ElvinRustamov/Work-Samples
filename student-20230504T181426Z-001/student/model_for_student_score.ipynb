**About Dataset:**
This data approach student achievement in secondary education of two Portuguese schools. The data attributes include student grades, demographic, social and school related features) and it was collected by using school reports and questionnaires. Two datasets are provided regarding the performance in two distinct subjects: Mathematics (mat) and Portuguese language (por). In [Cortez and Silva, 2008], the two datasets were modeled under binary/five-level classification and regression tasks. Important note: the target attribute G3 has a strong correlation with attributes G2 and G1. This occurs because G3 is the final year grade (issued at the 3rd period), while G1 and G2 correspond to the 1st and 2nd period grades. It is more difficult to predict G3 without G2 and G1, but such prediction is much more useful (see paper source for more details).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
student_math_dataset = pd.read_csv("/content/drive/MyDrive/Projects/student/student-mat.csv", delimiter=';')
```

```python
df = student_math_dataset.copy()
df.head()
```

```python
print(df.info())
print(20 * "*")
print(df.isnull().sum())
```

```python
df.drop(['school', 'guardian', 'address', 'famsize', 'Mjob', 'Fjob', 'reason'], inplace=True, axis=1)
df.info()
```

```python
# df.drop(list(df.loc[df['Fedu'] == 0].index) + list(df.loc[df['Medu'] == 0].index), axis=0, inplace=True)
df['Fedu'] = pd.Categorical(df['Fedu'], categories = [0, 1, 2, 3, 4, 5], ordered = True)
df['Medu'] = pd.Categorical(df['Medu'], categories = [0, 1, 2, 3, 4, 5], ordered = True)

df['traveltime'] = pd.Categorical(df['traveltime'], categories = [1, 2, 3, 4], ordered = True)
df['studytime'] = pd.Categorical(df['studytime'], categories = [1, 2, 3, 4], ordered = True)
```

```python
def convert_category_common_columns(dataframe, *args):
  for column in args:
    dataframe[column] = pd.Categorical(dataframe[column], categories = [1, 2, 3, 4, 5], ordered = True)
  return dataframe

convert_category_common_columns(df, 'freetime', 'famrel', 'goout', 'Dalc', 'Walc', 'health')
```

```python
df = df.replace('yes', 1)
df = df.replace('no', 0)
df = df.replace("F", 0)
df = df.replace("M", 1)
df = df.replace("T", 0)
df = df.replace("A", 0)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn import metrics
```

```python
X = df.drop(['G3'], axis=1)
y = df['G3']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 1)
```

```python
regression = LinearRegression()
regression.fit(X_train, y_train)
```

```python
y_pred = regression.predict(X_test)
pd.DataFrame({"Y hat": y_pred, "Real value":y_test})
```

#**Forecasting Errors**

Forecast Error determines the deviation between the forecast and the actual demand/sales.

Forecast Error measures are widely used in the following two scenarios:

- To evaluate the statistical forecast model fit by allowing several forecast models to compete against each other. This estimation is based on the ex-post forecast produced by forecast models against the actual sales
To evaluate the forecast performance by measuring the error between the forecasted sales and the actual sales, which most planners do in real time to report in their dashboards

Forecast Error measures can be classified into two groups:

- Percentage errors (or relative errors) - These are scale-independent (assuming the scale is based on quantity) by specifying the size of error in percentage and is easy to compare the forecast error between different data sets/series. Examples: Mean Percentage Error (MPE), Mean Absolute Percentage Error (MAPE), Weighted Mean Absolute Percentage Error (WMAPE) & Mean Absolute Squared Error (MASE).

- Scale-dependant errors (or absolute errors) – The size of error is measured in units by making it difficult to compare the forecast performance between different data sets/series. Examples: Mean Absolute Deviation (MAD), Mean Squared Error (MSE), Root Mean Square Error (RMSE), Error Total (ET) or Total Absolute Error (TAE).

```python
# MAE
print("Mean Absolute Error     : ", metrics.mean_absolute_error(y_test, y_pred))

# MSE
print("Mean Squared Error      : ", metrics.mean_squared_error(y_test, y_pred))

# RMSE
print("Root Mean Squared Error : ", np.sqrt(metrics.mean_squared_error(y_test, y_pred)))

#R2 Score
print("R2 Score                : ", metrics.r2_score(y_test, y_pred))
```
